```python
import csv
import sys
from serpapi import GoogleSearch
from google.colab import files  # Menambahkan import untuk download file di Colab

def main():
    api_key = "53d80b89a57976f12254eb0346f003f6a935615080910e48613ba7db7ad81dfc"  # Ganti dengan API Key Anda

    # Daftar keyword pencarian tambahan yang ingin dipakai
    keywords = [
        "tempat wisata ibadah di medan",
        "wisata religi medan",
        "religious site medan",
        "wisata Gereja Kristen di medan",
        "wisata gereja Katolik di medan",
        "wisata gereja Protestan di medan",
        "tempat wisata religi medan",
        "tempat wisata keagamaan medan",
        "Masjid wisata religi medan",
        "wisata religi Islam medan",
        "religious tourism medan",
        "tempat wisata suci medan",
        "wisata pura di medan"
    ]

    all_results = {}
    for q in keywords:
        print(f"Mencari dengan keyword: {q}")
        params = {
            "engine": "google_maps",
            "q": q,
            "type": "search",
            "api_key": api_key
        }
        search = GoogleSearch(params)
        try:
            results = search.get_dict()
        except Exception as e:
            print(f"Error memanggil SerpApi untuk '{q}': {e}")
            continue

        local_results = results.get("local_results", [])
        if not local_results:
            print(f"Tidak ada data untuk keyword: {q}")
            continue

        # Gabungkan hasil tanpa duplikasi berdasarkan place_id
        for item in local_results:
            pid = item.get("place_id", "")
            if pid and pid not in all_results:
                all_results[pid] = item

    if not all_results:
        print("Tidak ditemukan data untuk semua keyword.")
        sys.exit(1)

    # Header untuk file CSV
    header = [
        "ID", "Nama Tempat", "Alamat Lengkap", "Latitude", "Longitude",
        "Kategori", "Jam Buka", "Rating", "Jumlah Ulasan", "Nomor Telepon", "Website",
        "Link Google Maps"
    ]

    try:
        # Menyimpan data ke file CSV
        with open('/content/wisata_religi_medan.csv', 'w', newline='', encoding='utf-8') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow(header)

            for pid, item in all_results.items():
                id_tempat = pid
                nama_tempat = item.get("title", "")
                alamat = item.get("address", "")
                latitude = item.get("latitude", "")
                longitude = item.get("longitude", "")
                kategori = item.get("category", "")
                jam_buka = item.get("opening_hours", "")
                rating = item.get("rating", "")
                jumlah_ulasan = item.get("reviews", "")
                nomor_telepon = item.get("phone", "")
                website = item.get("website", "")
                link_maps = item.get("maps_link", "") or f"https://www.google.com/maps/place/?q=place_id:{pid}"

                csv_writer.writerow([
                    id_tempat, nama_tempat, alamat, latitude, longitude,
                    kategori, jam_buka, rating, jumlah_ulasan, nomor_telepon, website, link_maps
                ])

        print("Data semua keyword berhasil disimpan ke wisata_religi_medan.csv")

        # Menyediakan file untuk diunduh di Google Colab
        files.download('/content/wisata_religi_medan.csv')

    except Exception as e:
        print(f"Error saat menyimpan file CSV: {e}")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import string

df = pd.read_csv('wisata_religi_medan.csv')
print(df.head())
```

```python
df.drop(columns=['ID', 'Nomor Telepon', 'Kategori'], inplace=True, errors='ignore')
```

```python
def tentukan_kategori_dan_agama(nama):
    nama = str(nama).lower()

    # Daftar pengecualian berdasarkan nama unik
    pengecualian = {
        'graha bunda maria annai velangkanni': ('Gereja', 'Katolik'),
        'shri supramaniyam temple': ('Pura', 'Hindu'),
        'mariamman temple': ('Pura', 'Hindu'),
        'little india gate': ('Lainnya', 'Hindu'),
        'taman alam lumbini': ('Vihara', 'Buddha'),
        'pubbarama buddhist centre': ('Vihara', 'Buddha'),
        'raksa buana great temple': ('Pura', 'Hindu'),
    }

    for kunci in pengecualian:
        if kunci in nama:
            return pengecualian[kunci]

    # Kata kunci umum
    if any(k in nama for k in ['masjid', 'mushola', 'mosque']):
        return 'Masjid', 'Islam'
    elif any(k in nama for k in ['gereja', 'church', 'katolik', 'protestan', 'gkpi', 'gbi', 'gpib', 'gki', 'gloria']):
        return 'Gereja', 'Kristen'
    elif any(k in nama for k in ['vihara', 'wihara', 'buddha', 'buddhist']):
        return 'Vihara', 'Buddha'
    elif any(k in nama for k in ['pura', 'hindu', 'temple']):
        return 'Pura', 'Hindu'
    elif any(k in nama for k in ['klenteng', 'konghucu', 'kelenteng']):
        return 'Klenteng', 'Konghucu'
    elif any(k in nama for k in ['makam', 'ziarah', 'mausoleum']):
        return 'Makam Ziarah', 'Islam'
    else:
        return 'Lainnya', 'Tidak Diketahui'


df[['Kategori', 'Agama']] = df['Nama Tempat'].apply(lambda x: pd.Series(tentukan_kategori_dan_agama(x)))
```

```python
kolom_awal = list(df.columns)
if 'Rating' in kolom_awal:
    rating_index = kolom_awal.index('Rating')
    kolom_awal.remove('Kategori')
    kolom_awal.remove('Agama')
    kolom_awal.insert(rating_index, 'Agama')
    kolom_awal.insert(rating_index, 'Kategori')
    df = df[kolom_awal]
```

```python
df.to_csv("wisata_religi_medan_2.csv", index=False)
```
